```python
import cv2

tracker = cv2.TrackerKCF_create()

cap = cv2.VideoCapture("../Computer_Vision_Master/race.mp4")

ret,frame = cap.read()

bbox = cv2.selectROI(frame)
#print(bbox)

ok = tracker.init(frame,bbox)
#print(ok)

while True:
    
    ok,frame = cap.read()
    #print(ok)
    
    if not ok:
        break
        
    ok, bbox = tracker.update(frame)
    #print(bbox)
    #print(ok)
    
    if ok:
        
        (x,y,w,h) = [int(v) for v in bbox]
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0),2,1)
        
    else:
        
        cv2.putText(frame, "Error", (100,800), cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,0,255), 2 )
        
    cv2.imshow("Tracking",frame)
    
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
