#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
img = cv2.imread(path)
fix_img = cv2.cvtColor(src = img , code = cv2.COLOR_BGR2RGB)

vector = np.array([[400,400], [200,700], [600,700]])
#print(vector.shape)
pts = vector.reshape((-1,1,2))
#print(pts.shape)

tri_img = cv2.polylines(img = fix_img,pts = [pts], isClosed = True , color = (0,0,255), thickness = 20)
plt.imshow(tri_img)
plt.show()
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import cv2
import numpy as np

def draw_circle(event,x,y,flags,param):
    
    if event == cv2.EVENT_RBUTTONDOWN:
        
        cv2.circle(img = img, center = (x,y), radius = 100, color = (0,0,255), thickness = 10)

img = cv2.imread("DATA/dog_backpack.jpg")
        
cv2.namedWindow(winname = "dog")

cv2.setMouseCallback("dog", draw_circle)

while True:
    
    cv2.imshow(winname = "dog",mat = img)
    
    if cv2.waitKey(delay = 20) & 0xFF == ord("q"):
        break
        
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

drawing = False
ix,iy = -1,-1

# Aslında ben burda ix,iy güncelliyorum yani tıkladığım yer x,y oluyor !!!!

def draw_rectangle(event,x,y,flags,param):
    
    global drawing,ix,iy
    
    
    # İlk önce bir ix,iy yeri belirliyoruz tıkladıktan sonra 
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        drawing = True
        ix,iy = x,y
    
    # Sonra bizim yeni x,y ler otomatik olarak belirlenecek
    # ix,iy zaten tıkladığımda belirlendi bunlar hafızada
    # x,y ler yeni değerlerini alıcak hareket edincede bu değerler değişicek
    
    elif event == cv2.EVENT_MOUSEMOVE:
        
        if drawing == True:
            
            cv2.rectangle(img = img, pt1 = (ix,iy), pt2 = (x,y), color = (255,0,0), thickness = -1)
    
    # Burda ise tıklamayı bırakınca çizimden çıkıcak ve son halini gösterecek
    
    elif event == cv2.EVENT_LBUTTONUP:
        
        drawing = False
        
        cv2.rectangle(img = img, pt1 = (ix,iy), pt2 = (x,y), color = (255,255,0), thickness = -1)
        
img = np.zeros((512,512,3))

cv2.namedWindow("rec")
cv2.setMouseCallback("rec",draw_rectangle)
        
while True:
    
    cv2.imshow(winname = "rec",mat = img)
    
    if cv2.waitKey(delay = 1) & 0xFF == ord("q"):
        break
        
cv2.destroyAllWindows()
```
