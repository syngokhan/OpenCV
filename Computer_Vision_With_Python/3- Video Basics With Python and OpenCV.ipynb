```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Windows --> DIVX
# Linux or MacOs --> XVID

writer = cv2.VideoWriter("mysupervideo.mp4",cv2.VideoWriter_fourcc(*"XVID"),30,(width,height))

while True:
    
    ret,frame = cap.read()
    
    writer.write(frame)
    
    #gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    #cv2.imshow(winname = "Frame",mat = gray)
    
    cv2.imshow(winname = "Frame", mat = frame)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        
        break
        
cap.release()
writer.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

x = width // 2
y = height // 2

w = width // 4
h = height // 4
 
while True:
    
    ret, frame = cap.read()
    #print(width,height)
    
    cv2.rectangle(frame,(x,y),(x+w, y+h),(0,0,255),thickness = 4)
    
    cv2.imshow("Frame",frame)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Linux or MacOs --- > XVID
# writer = cv2.VideoWriter("myvideo.mp4",cv2.VideoWriter_fourcc(*"XVID",30,(width,height)))

## CALLBACK FUNCTION RECTANGLE

def draw_rectangle(event,x,y,flags,param):
    
    global pt1,pt2,topLeft_Clicked,botRight_Clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        if topLeft_Clicked == True and botRight_Clicked == True:
            
            pt1 = (0,0)
            pt2 = (0,0)
            topLeft_Clicked = False
            botRight_Clicked = False
        
        # Burda şunu düşünme zaten ikiside False Hemen ikiside aynı değer alıcak gibi düşünme
        # Sırası ile gider yani ilk tıkladığında topLeft girer ikinci tıkladığında ise botRight girer devreye
        
        if topLeft_Clicked == False:
            
            pt1 = (x,y)
            topLeft_Clicked = True
            
        elif botRight_Clicked == False:
        
            pt2 = (x,y)
            botRight_Clicked = True
            
## GLOBAL VARIABLES

pt1 = (0,0)
pt2 = (0,0)
topLeft_Clicked = False
botRight_Clicked = False

## CONNECT TO THE CALLBACK

cap = cv2.VideoCapture(0)

cv2.namedWindow("Test")
cv2.setMouseCallback("Test",draw_rectangle)


while True:
    
    ret,frame = cap.read()
    
    #writer.write(frame)
    
    if topLeft_Clicked:
        
        cv2.circle(frame,pt1,5,(0,0,255),thickness = -1)
    
    #print(pt1,pt2)
    
    if topLeft_Clicked and botRight_Clicked:
        
        cv2.rectangle(frame,pt1,pt2,(0,0,255),3)
        
        
    cv2.imshow("Test",frame)
        
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
        
        
cap.release()
cv2.destroyAllWindows()
```
